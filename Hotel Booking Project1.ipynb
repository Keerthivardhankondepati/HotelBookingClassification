#### Handling Outliers of the countinous variables using IQR-distance from Median

```python
# adr outliers are handled 
Q1 = df['adr'].quantile(0.25)
Q3 = df['adr'].quantile(0.75)
IQR = Q3 - Q1    #IQR is interquartile range. 

filter_adr = (df['adr'] >= Q1 - 1.5 * IQR) & (df['adr'] <= Q3 + 1.5 *IQR)
print('Total Outliers detected in the feature adr is %d' %(len(filter_adr)- len(df.loc[filter_adr])))
```

```python
new_df = df.copy()
new_df['adr'].where(filter_adr,inplace= True)
mean_value = new_df['adr'].mean()
new_df['adr'].fillna(value=mean_value, inplace=True)
new_df.shape
```

```python
# lead_time outliers are handled 
Q1 = df['lead_time'].quantile(0.25)
Q3 = df['lead_time'].quantile(0.75)
IQR = Q3 - Q1    #IQR is interquartile range. 

filter_lead = (df['lead_time'] >= Q1 - 1.5 * IQR) & (df['lead_time'] <= Q3 + 1.5 *IQR)
print('Total Outliers detected in the feature adr is %d' %(len(filter_lead)- len(df.loc[filter_lead])))
```

```python
new_df['lead_time'].where(filter_lead,inplace= True)
mean_value = new_df['lead_time'].mean()
new_df['lead_time'].fillna(value=mean_value, inplace=True)
```

```python
#Box plots after handling outliers using IQR approach
fig, axes = plt.subplots(1,2,figsize = (16,10))
axes = axes.flatten()
cols_to_plot = ['adr','lead_time']

for i in range(0,len(cols_to_plot)):
    sns.boxplot(x="is_canceled", y= new_df.loc[:,cols_to_plot[i]], data=df, orient='v', ax=axes[i])

plt.tight_layout()
plt.show()
```

```python
# Create a function to display information about the data
pd.set_option('display.max_columns', 50)
def display_data_info(data):

    # Print the number of null values in the data and the percentage of null values
    print("FEATURES WITH NULL VALUES\n")
    rows=data.shape[0]
    for col in data.columns:
    
        null_values = data[col].isnull().sum() # Calculate the number of null values
        

    # If the null values > 0 print the column, number of null values and the percentage of null values
        if null_values > 0:
            perc_null_vals = float(null_values)*100/rows
            print("FEATURE: {}".format(col))
            print("NULL VALUES COUNT: {}".format(null_values))
            print("PERCENTAGE VALUES COUNT: {:.3f}%".format(perc_null_vals))
            print()
```

```python
display_data_info(new_df) # we can see that the below features are having null values
```

## Q6 ML approaches and algorithms

here the problem statement is to find out if the reseravation will be cancelled or not in fututre after a booking is done.
this is a classification problem. Hence we have deiced to use the classification techniques like Logistic regression , decision trees



```python
from sklearn.model_selection import train_test_split
y = new1_df['is_canceled']
X = new1_df.drop('is_canceled', axis = 1)
X_train, X_test, y_train, y_test = train_test_split(X,y,random_state=51,test_size=0.3)
```

```python
from sklearn.linear_model import LogisticRegression
logreg = LogisticRegression().fit(X_train, y_train)
y_pred_log = logreg.predict(X_test)
```

```python
from sklearn.metrics import accuracy_score, confusion_matrix, classification_report
acc_lg = accuracy_score(y_test, y_pred_log)
conf = confusion_matrix(y_test, y_pred_log)
clf_report = classification_report(y_test, y_pred_log)
```

```python
from sklearn.metrics import plot_confusion_matrix
plot_confusion_matrix(logreg, X_test, y_test)

print("The accuracy of the Logistic regression is:",acc_lg)
print("\n")
plt.show()
print("Evaluation Metrics Logistic Regression")
print(clf_report)
```

#### Decsion Tree Classifier

```python
from sklearn.tree import DecisionTreeClassifier
dtc = DecisionTreeClassifier()
dtc.fit(X_train, y_train)

y_pred_dtc = dtc.predict(X_test)
```

```python
from sklearn.metrics import accuracy_score, confusion_matrix, classification_report
acc_dtc = accuracy_score(y_test, y_pred_dtc)
conf = confusion_matrix(y_test, y_pred_dtc)
clf_report = classification_report(y_test, y_pred_dtc)
acc_dtc
```

```python

print("The accuracy of the Decision Tree is:",acc_dtc)
print("\n")
plt.show()
print("Evaluation Metrics Decision Tree")
print(clf_report)
```

```python
import imblearn
from imblearn.over_sampling import RandomOverSampler
over = RandomOverSampler()
x_train, y_train = over.fit_resample(X_train, y_train)
df_sampled_final = pd.DataFrame(x_train, columns = X_train.columns)
sns.countplot(y_train)
```

#### after balancing the imbalanced bata from the dataset there is a considerable inseracse in the accuarcy of the test prediction and Recall which is a good sign as there will be very few occurances where a booking might be predicted as will not be cancelled which is_cancelled in reality



```python
from sklearn.ensemble import RandomForestClassifier
rf = RandomForestClassifier()
rf1 = rf.fit(x_train, y_train)
y_pred_rf = rf1.predict(X_test)

from sklearn.metrics import accuracy_score, confusion_matrix, classification_report
acc_rf = accuracy_score(y_test, y_pred_rf)
conf_rf = confusion_matrix(y_test, y_pred_rf)
clf_report_rf = classification_report(y_test, y_pred_rf)
print("Accuracy of Random Forest model : ",acc_rf)
print("consution matrix:",conf_rf)
print(clf_report_rf)
```
